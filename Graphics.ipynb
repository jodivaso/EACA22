```python
import pandas as pd
import seaborn as sns
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

## Graphics for benchmarks_kz3_kzp6

```python
df = pd.read_csv("benchmarks_kz3_kzp6.txt",sep=";")

%matplotlib inline
x = df["Dimension"]
y1 = df["Time_parallelism_columns"]
y2 = df["Time_parallelism_basic"]
y3 = df["Time_no_parallelism"]
fig, ax1 = plt.subplots(1, 1, figsize=(10,10))
      
ax1.plot(x, y1, label='Parallelism by columns')
ax1.plot(x, y2, label='Parallelism by matrix')
ax1.plot(x, y3, label='No parallelism')
ax1.set_yscale("symlog", base=2) #the log transformation
#ax1.set_yscale('symlog', linthresh=0.01)
ax1.set_title("Benchmark for XXX");
ax1.set_xlabel("Dimension");
ax1.set_ylabel("Time (seconds)");
ax1.set_yticks([0.1,1,10, 50, 200, 500,1000, 2000, 6000])
ax1.get_yaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
handles, labels = ax1.get_legend_handles_labels()
ax1.legend()
plt.show()
```
